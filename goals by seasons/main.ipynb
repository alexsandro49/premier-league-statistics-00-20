```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../results.csv', encoding='ISO-8859-1')
df_00_to_21 = df[df['Season'] >= '2000-01']
```

```python
df_arsenal_by_season = df_00_to_21[df_00_to_21['HomeTeam'] == 'Arsenal'].groupby('Season').sum().reset_index()
df_arsenal_by_season['TGS'] = df_arsenal_by_season['FTHG'] + df_00_to_21[df_00_to_21['AwayTeam'] == 'Arsenal'].groupby('Season').sum().reset_index()['FTAG']
df_arsenal_by_season['TGC'] = df_arsenal_by_season['FTAG'] + df_00_to_21[df_00_to_21['AwayTeam'] == 'Arsenal'].groupby('Season').sum().reset_index()['FTHG']

plt.bar(df_arsenal_by_season['Season'], df_arsenal_by_season['TGS'], label='Total goals scored', color='#002A74ff')
plt.bar(df_arsenal_by_season['Season'], df_arsenal_by_season['TGC'], label='Total goals conceded', color='#DB0000ff')
plt.title('Arsenal goals by season')
plt.xlabel('Seasons')
plt.ylabel('Goals')
plt.xticks([df_arsenal_by_season.loc[i, 'Season'] for i in range(0, len(df_arsenal_by_season), 4)])
plt.yticks(np.arange(10, 91, 10), minor=True)
plt.legend(loc='lower right')
plt.savefig('arsenal_goals_by_season.png')
plt.show()
```
